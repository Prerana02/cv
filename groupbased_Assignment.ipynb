```python
import cv2
import numpy as np


# Load an image
image = cv2.imread('face.jpeg')
# Translation
tx = 50  # Translation in the x-axis
ty = 30  # Translation in the y-axis
translation_matrix = np.float32([[1, 0, tx], [0, 1, ty]])
translated_image = cv2.warpAffine(image, translation_matrix, (image.shape[1], image.shape[0]))

# Rotation
angle = 30  # Rotation angle in degrees
rotation_matrix = cv2.getRotationMatrix2D((image.shape[1] / 2, image.shape[0] / 2), angle, 1)
rotated_image = cv2.warpAffine(image, rotation_matrix, (image.shape[1], image.shape[0]))

# Scaling
scale_x = 1.5  # Scale factor in x-axis
scale_y = 0.5  # Scale factor in y-axis
scaling_matrix = np.float32([[scale_x, 0, 0], [0, scale_y, 0]])
scaled_image = cv2.warpAffine(image, scaling_matrix, (image.shape[1], image.shape[0]))
# Shearing
shear_x = 0.2  # Shear factor in x-axis
shear_y = 0.1  # Shear factor in y-axis
shearing_matrix = np.float32([[1, shear_x, 0], [shear_y, 1, 0]])
sheared_image = cv2.warpAffine(image, shearing_matrix, (image.shape[1], image.shape[0]))

from google.colab.patches import cv2_imshow

# Display the original and transformed images
cv2_imshow(image)
# You can add labels or use print statements to label the images
print("Original Image")

cv2_imshow(translated_image)
print("Translated Image")

cv2_imshow(rotated_image)
print("Rotated Image")

cv2_imshow(scaled_image)
print("Scaled Image")

cv2_imshow(sheared_image)
print("Sheared Image")

cv2.waitKey(0)
cv2.destroyAllWindows()


```
